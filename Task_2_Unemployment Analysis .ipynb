##Importing Labries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
import plotly.express as px
```

##Importing Dataset

```python
df= pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
print(df)
```

##Average Unemployment Rate

```python
print('Average Unemployment Rate: ', df[' Estimated Unemployment Rate (%)'].mean())
```

##Total no of Employed People

```python
print('Total no of Employed People: ', df[' Estimated Employed'].sum())
```

##Total no of Employed People for Indusial States

```python
d_f = df.groupby('Region')[' Estimated Employed'].sum().sort_values(ascending=False)
d_f
```

##Average Unemployment Rate for Indusial States

```python
groups = df.groupby('Region')[' Estimated Unemployment Rate (%)']
for region, data in groups:
    print(f"Region: {region}")
    print(data.mean())
    print('\n')
```

```python
df['Region'].value_counts()
```

#Visually Analysing Dataset



##Average Unemployment Rate on for Indusial States

```python
avg_unemp_state = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(10, 6))
avg_unemp_state.plot(kind='bar', color='orange')
plt.title('Average Unemployment Rate on for Indusial States')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
heatmap = df[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)', 'Region', 'longitude', 'latitude']]
heatmap = heatmap.corr()
plt.figure(figsize=(8,8))
sn.set_context('notebook', font_scale=1)
sn.heatmap(data=heatmap, annot=True, cmap=sn.cubehelix_palette(as_cmap=True))
```
